```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('/Users/brinkley97/Documents/development/random_code/linear_regression_data/train.csv')
df_test = pd.read_csv('/Users/brinkley97/Documents/development/random_code/linear_regression_data/test.csv')

# i believe x & y coming from file headers (or columns)
x_train = df_train['x']
y_train = df_train['y']

x_test = df_test['x']
y_test = df_test['y']

# take values & store in an array
x_train = np.array(x_train)
y_train = np.array(y_train)

x_test = np.array(x_test)
y_test = np.array(y_test)


# reshape : gives a new shape to an array without changing its data
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
# number of total data points - see under "About this file" 
    # here https://www.kaggle.com/andonians/random-linear-regression/version/2?select=train.csv
n = 700

# learning rate
alpha = 0.0001

# return a new array of given shape and type, filled with zeros
    # 700 rows x 1 column
a_0 = np.zeros((n, 1))
# a_0 = a_0.T
# print(a_0)

a_1 = np.zeros((n, 1))


epochs = 0
while(epochs < 1000) :
#     linear equation : look at image above
    y = a_0 + a_1 * x_train
    
    error = y - y_train
#     print(error)
    
#     squared is **2 but try np.square() as well
    mean_sq_er = np.sum(error ** 2)
#     print("mse : ", mean_sq_er)
    
    mean_sq_er = mean_sq_er / n
#     print("mse : ", mean_sq_er)
    
#     update a_0
    a_0 = a_0 - alpha * 2 * np.sum(error) / n
    
#     update a_1
    a_1 = a_1 - alpha * 2 * np.sum(error * x_train) / n 
    
    epochs += 1
    
    if(epochs%10 == 0) :
        print("mse : ", mean_sq_er)
```
