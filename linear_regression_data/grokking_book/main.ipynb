```python
def square_trick(base_price, price_per_room, number_of_rooms, actual_price, learning_rate):
    '''Calculate the predicted price
    
    Arguments:
    base_price -- int
    price_per_room -- int
    number_of_rooms -- int
    actual_price -- int
    learning_rate -- int
    
    '''
    predicted_price = base_price + (price_per_room * number_of_rooms)
    # print("predicted_price : base_price + price_per_room * number_of_rooms = predicted_price")
    # print("predicted_price :    ", base_price, "   +     ", price_per_room, "       *        ", number_of_rooms, "       =      ", predicted_price)
    error = actual_price - predicted_price
    # print("\nerror : actual_price - predicted_price = error")
    # print("error :    ", actual_price, "     -       ", predicted_price, "   =   ", error)
    base_price = base_price + (learning_rate * error) 
    price_per_room = price_per_room + learning_rate * number_of_rooms * error
    
    return price_per_room, base_price
    # return predicted_price, error, base_price, price_per_room
```

1. slope [measure of how steep the line is]
    - $ m = rise / run $
    - per 1 feature (x, run), how much does my label (y, rise) changes?
2. y-int [of a line, it's the height @ whichh the line crosses the y-axis]
    - $ b = (x, y) $, where x = 0, then y = ? 
    - if line doesn't cross the y-axis, then there does NOT exists a y-int
3. linear equation [equation of a line]
    - $ y = m * x + b $, where
        - x (features), y (labels) are points
        - m and b are parameters

```python
def linear_regression(features, labels, learning_rate, epochs):
    '''Find the best line to fit the data
    
    Arguments:
    features -- list
    labels -- list
    learning_rate -- float
    epochs -- int
    
    Return:
    tuple of price_per_room -- int and base_price -- int
    '''
    
    # Generate random data
    # slope 
    price_per_room = np.random.random()
    # price_per_room = 50
    # print(price_per_room)
    
    # y_int
    base_price = np.random.random()
    # base_price = 100
    # print(base_price)
    
    for epoch in range(epochs):
        if True:
            # draw the line without points
            utils.draw_line(price_per_room, base_price, starting=0, ending=8)
            
        # idx_of_observation = np.random.randint(0, len(features)-1)
        idx_of_observation = np.random.randint(1, len(features))
        # print("idx_of_observation : ", idx_of_observation)
        
        # for each iteration (up to the max), let's get a random observation from the features and prices
        num_rooms = features[idx_of_observation]
        price = labels[idx_of_observation]
        # print("\n@ index", idx_of_observation, "has", num_rooms, "rooms for $", price)
        
        # update the price per room and base price
        price_per_room, base_price = square_trick(base_price, price_per_room, num_rooms, price, learning_rate)
        # print("\n", price_per_room, base_price)
        
    utils.draw_line(price_per_room, base_price, 'black', starting=0, ending=8)
    utils.plot_points(features, labels)
    # print('Price per room:', price_per_room)
    # print('Base price:', base_price)
    return price_per_room, base_price
```
