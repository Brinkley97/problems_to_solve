```python
#unit step signal
#variables => time, amplitude, shift
    #time => point on the graph &/or x-axis; u(t) or u(-t)
    #amplitude => up 5u(t) & down 0.5u(t)
    #shift from origin (t=0) => right : u(t - 5) & left u(-t - 5)
    #therefore, u = f(t, A, Shift) => u is the function of t, A, Shift

#import* = import all packages
from numpy import*
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#linspace => generates linearly spaced values
    #y = linspace(x1,x2,n) generates n points
    #n => for accuracy; lenth
t = linspace(-10, 10, 1000)
#lenght equals the n value for variable t
n = len(t)
a = 1
shift = 0
#create array of 0s
u = zeros(n)
#straight line
# plt.plot(t, u)

#add step portion to straight line
for i in range(0, n) :
    if t[i] >= -shift :
#         print(-shift)
        u[i] = a
plt.plot(t, u)
plt.grid()
```

```python
#n differs from above to show it's contribution
t2 = linspace(-10, 10, 10)
n = len(t2)
a = 1
shift = 0
u = zeros(n)
for i in range(0, n) :
    if t2[i] >= -shift :
        u[i] = a
plt.plot(t2, u)
plt.grid()
```

```python
#n differs from above to show it's contribution
t3 = linspace(-10, 10, 1000)
n = len(t3)
a = 0
shift = 0
u = ones(n)
for i in range(0, n) :
    if t3[i] >= -shift :
        u[i] = a
plt.plot(t3, u)
plt.grid()
```

```python
#function for unit step signal
def function_step(t, a, shift) :
    for i in range(0, n) :
        if t[i] >= -shift :
            u[i] = a
    return u
```

```python
t = linspace(-10, 10, 1000)
n = len(t)
u = zeros(n)
#change of a & shift
u = function_step(t, 5, 3)
plt.plot(t, u)
plt.grid()
```

```python
t = linspace(-10, 10, 1000)
n = len(t)
u = zeros(n)
#change of a & shift
u = function_step(t, 5, -3)
plt.plot(t, u)
plt.grid()
```
