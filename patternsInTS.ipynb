```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 3, figsize = (20, 4), dpi = 100)
pd.read_csv(
            'https://raw.githubusercontent.com/selva86/datasets/master/guinearice.csv',
            parse_dates=['date'],
            index_col='date'
            ).plot(title='Trend Only', legend=False, ax=axes[0])

pd.read_csv(
            'https://raw.githubusercontent.com/selva86/datasets/master/sunspotarea.csv', 
            parse_dates=['date'],
            index_col='date'
            ).plot(title='Seasonality Only', legend=False, ax=axes[1])

pd.read_csv(
            'https://raw.githubusercontent.com/selva86/datasets/master/AirPassengers.csv', 
            parse_dates=['date'],
            index_col='date'
            ).plot(title='Trend and Seasonality', legend=False, ax=axes[2])
```
