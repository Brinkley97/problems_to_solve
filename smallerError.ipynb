# 2.3.6 Regularization
1. Which gives a smaller error


```python
import numpy as np
```

```python
def solution1(x1, x2):
    outcome1 = x1 + x2
    return outcome1

def solution2(x1, x2):
    outcome2 = (10 * x1) + (10 * x2)
    return outcome2

def sigmoid(x):
    sig = 1 / (1 + np.exp(-x))
    return sig
```

```python
a = solution1(1,1)
b = solution1(-1,-1)
c = solution2(1,1)
d = solution2(-1,-1)
```

```python
print(a)
print(b)
print(c)
print(d)
```

```python
sigA = sigmoid(a)
sigB = sigmoid(b)
sigC = sigmoid(c)
sigD = sigmoid(d)
```

```python
print(sigA)
print(sigB)
print(sigC)
print(sigD)
```

1. Apply [Quick Sort Algorithm](http://localhost:8888/notebooks/random_code/quicksortAlgorithm.ipynb) to determine order

```python
solution1Answer = sigB - sigA
solution2Answer = sigD - sigC
```

```python
print(solution1Answer, solution2Answer)
```
