## 1.1.1 (Dot Product)


```python
def my_dot_product(x, y) :
    c = []
    print(x[0], y[0][0])
    a11 = x[0] * y[0][0]
    print("a11 : ", a11)
    
    print(x[1], y[0][0])
    a21 = x[1] * y[0][0]
    print("a21 : ", a21)
    
    print(x[0], y[0][1])
    a12 = x[0] * y[0][1]
    print("a12 : ", a12)
    
    print(x[1], y[0][1])
    a22 = x[1] * y[0][1]
    print("a22 : ", a22)
            
    c.append(a11)
    c.append(a21)
    c.append(a12)
    c.append(a22)
    
    return c
```

```python
x = np.array([[1],
              [2]
             ])
y = np.array([[2, 3]])
```

```python
my_dot_product(x, y)
```

```python
def np_dot_product(x, y, n) :
    c = np.dot(x, y)
    
    return c
```

```python
np_dot_product(x, y, n)
```
